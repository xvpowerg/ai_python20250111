## 合併

```python
import pandas as pd
d = {'one':pd.Series([1,2,3],index=['a','b','c']),
     'two':pd.Series([1,2,3,4],index=['a','b','c','d'])}
df1 = pd.DataFrame(d)
df1["three"] = [10,11,12,13]
print(df1)
df2 = pd.DataFrame([[20, 21, 22],
                    [23, 24, 25],
                    [26,27, 28]], columns=['one', 'two', 'three'],index=['a','b','c'])
print(df2)
print("=================")
df3 = pd.concat([df1,df2],ignore_index=True)#軸0方向合併
df3.index =['a','b','c',"d","e","f","g"]
print(df3)
df4 = pd.concat([df1,df2],ignore_index = True,axis=1)#軸1方向合併
df4.columns = ['a','b','c',"d","e","f"]
print(df4)
```

### 使用 loc(文字 m:n 會包含到n)
+ "location" 的縮寫
+ .loc[ ], 中括弧裡面是先 row 後 column，以逗號分開。
+ row 與 column 要放入的是 row 標籤與 column 標籤。
+ 可於 row 或 column 加入「：」 ，代表指定起訖範圍。

```python
import  pandas as pd
import  numpy as np
df=pd.DataFrame(
    np.arange(48).reshape(6,8),
    index=['a','b','c','d','e','f'],
    columns=['s','t','u','v','w','x','y','z'])
print(df)
print("===============")
print(df.loc["b",:])
print("===============")
print(df.loc["b","w"])
print("===============")
print(df.loc[["c","e"],["w","y"]])
print("===============")
print(df.loc["c":"e","w":"y"])
print("===============")
print(df.loc["b":,:"w"])
```

### in 與 not in

+ Pandas 有 isin 方法，代表如果資料存在於 isin 方法內的比對對象，那就傳回 True。
+ 如果比對對象是一個 dictionary：
  + 首先判斷對應的 index 是否存在。
  + 如果存在，接著判斷對應的位置是否存在。
  + 如果不存在，那麼全部為 False。
+ 如果是一個 DataFrame：
  + 首先就是 column 名稱要存在。
  + 原資料行列位置和對應的行列位置一一匹配，才返回 TRUE。
+ 若要設計一個 not in 的分析語法，請在函數方法前面加個 ~ 即可。

```python
import  pandas as pd
data_url='http://bit.ly/2cLzoxH'
gap1=pd.read_csv(data_url)
newData = gap1[gap1["year"].isin([1997,2007])]
print(newData)
newGap = gap1.query("year == 2007 or year == 1997")
print(newGap)
```

## 處理日期資訊
+ 可使用 pandas 的 date_range( ) 函數，可以建立日期資料。
+ 默認情況下，範圍的週期單位為天。
+ 若要表達出工作天 (不包含周六周日)，則請選擇 bdate_range( ) 方法。
+ date_range 或 bdate_range( ) 內的參數：
  + ('起始日期', periods=天數)
  + periods 由開始日取幾天
  + ('起始日期','結束日期')
  + 可加入另外一個參數：freq 代表範圍的週期單位
  
  https://pandas.pydata.org/docs/user_guide/timeseries.html

```python
import pandas as pd
print('日期請調整為上課時的時間')
a=pd.date_range('2025/04/05',periods=10)
print(a)
print('-'*70)
a=pd.date_range('2025/04/05',periods=10,freq='W')
print(a)
print('-'*70)
a=pd.date_range('2025/04/05',periods=10,freq='W-Mon')#每週一
print(a)
print('-'*70)
```

## 丟棄遺失值
+ 您可以使用 dropna 方法再搭配 axis 參數方式進行搭配。
+ 默認情況下，預設為 axis = 0，也就是會沿著 Rows 進行，當發現到任何值為 NA 就會整個 Row 刪除。
+ 若加入參數 how=‘all’ 代表整個 Row 資料都是遺失值情況下才可以刪除丟棄。
+ 加入參數 thresh=N 代表刪除包含少於 N 個觀察值的 row。

```python
import pandas as pd
import  numpy as np
from io import  StringIO
csv_data='''
A,B,C,D,E
2,3,4,5,
6,34,6,
10,,11,8,
,,,
3,,3,,
,5,,,
'''
drop=pd.read_csv(StringIO(csv_data))
print(drop)

drop1 = drop.dropna(axis=0)
print(drop1)
print("==================")
drop2 = drop.dropna(axis=1)
print(drop2)
print("==================")
drop3 = drop.dropna(axis=0,how="all")# 所有都是Nan才移除
print(drop3)
print("==================")
drop4 = drop.dropna(thresh=3)#至少要有 thresh 個有效資料才保留
print(drop4)
print("==================")
drop5 = drop.dropna(subset=["C","D"])#c跟d欄位都不是NaN的保留
print(drop5)
print("==================")
drop6 = drop.dropna(thresh=3,axis=1)#至少要有 thresh 個有效資料才保留
print(drop6)
```

## 填充遺失值
+ 可使用 fillna 函數進行填充
  + 輸入要填充的數值
    + 補上中位數：建議補上中位數而非平均值，這樣相對來說不會受到極端值的影響。
    + 根據原本的資料分布補上亂數。
+ 以下兩種方式可以向後或者向前進行填充
   + pad/ffill          代表向後填充
   + bfill/backfill   代表向前填充

```python
import pandas as pd
import  numpy as np
from io import  StringIO
csv_data='''
A,B,C,D,E
2,3,4,5,
6,34,6,
10,,11,8,
,,,
3,,3,,
,5,,,
'''
fill=pd.read_csv(StringIO(csv_data))
print(fill)
print("===============")
fill_1 = fill.fillna(0)
print(fill_1)
print("===============")
fill_2 = fill.fillna(0,limit=2)#只填上2個0
print(fill_2)
print("===============")
fill_3 = fill.ffill()## 如果當前欄位是NaN  向上一筆同欄直到非NaN的值 寫入目前欄位
print(fill_3)
print("===============")
fill_4 = fill.bfill()#如果當前欄位是NaN  向下一筆同欄直到非NaN的值 寫入目前欄位
print(fill_4)
```

```python
import pandas as pd
import  numpy as np
from io import  StringIO
csv_data='''
A,B,C,D,E
2,3,4,5,
6,34,6,
10,,11,8,
,,,
3,,3,,
,5,,,
'''
fill=pd.read_csv(StringIO(csv_data))
print(fill)
fillmean = fill.mean()
#print(fillmean)
for x in fill:
    col_to_fill = fill[x]
    col_to_fill.fillna(fillmean[x],inplace=True)
print(fill)
```
