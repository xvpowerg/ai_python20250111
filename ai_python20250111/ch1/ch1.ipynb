## 計算

```python
a,b = 10,3
ans = a + b
print(ans)
ans = a - b
print(ans)
ans = a * b
print(ans)
ans = a / b
print(ans)
ans = a // b #無條件捨去
print(ans)
ans = a / b
print(round(ans,2))#會四捨五入 取道小數點第2位
##餘數
ans = 9 % 2
print(ans)
```

```python
n = 2
x = 3
ans = x ** n #x的n次方
print(ans)

n = -1
x = 2
ans = x ** n
print(ans)

n = 0.5
x = 9
ans = x ** n
print(ans)
```

```python
n = 1/ 3
x = 64
print(n)
ans = x ** n
print(round(ans,2))
```

```python
ans = 10 + 2
print(ans,type(ans))
ans = 10 + 2.0 #計算式只要有浮點數 結果就是浮點數
print(ans,type(ans))
```
