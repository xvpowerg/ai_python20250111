## find

字串可以搜尋特定內容的位置。

搜尋的字串位置編號由 0 開始。

字串.find(搜尋字串,起始位置,結束位置)

搜尋時不包含結束位置。

若搜尋不到則回傳 -1。

若沒有起始位置，也沒有結束位置，預設由 0 開始。

若沒有結束位置，預設到最後一個字。

```python
str1 = "This is Python.That is Java,This is SQLite"
print(str1)
search1 = "Python"
index = str1.find(search1)
print(index)
search1 = "not"
index = str1.find(search1)#找不到回傳-1
print(index)
search1 = "t"
index = str1.find(search1)
print(index)
search1 = "t"
index = str1.find(search1,index + 1)#由哪個index開始找
print(index)
```

## join

join( ) : 於字串或著字元之間加入指定的文字

指定文字.join(要插入的字串)

```python
str1 = "-"
str2 = ("A","B","C")#tuple
print(str2)
print(type(str2))
newStr = str1.join(str2)
print(newStr)
print(type(newStr))
str3 = "Python"
newStr2 = ":".join(str3)
print(newStr2)
```
