```python
#
#
#add minus mult div
# v1 v2
#type = "add"

def calculate(v1,v2,type="add"):
    if type == "add":
        return v1 + v2
    elif type == "minus":
        return v1 - v2
    elif type == "mult":
        return v1 * v2
    elif type == "div":
        return v1 / v2
    else:
        print("Error!")
        return -1
        
ans =  calculate(6,2)
print(ans)
ans =  calculate(6,2,"minus")
print(ans)
ans =  calculate(6,2,"mult")
print(ans)
ans =  calculate(6,2,"div")
print(ans)    
```

```python
def sumValues(*args):
    v = 0
    for a in args:        
        v += a               
    return v    
print(sumValues())
print(sumValues(2))
print(sumValues(2,3))
print(sumValues(2,3,4))
print(sumValues(2,3,4,2))
```

```python
def calculate(*values,**args):
    cType = "+"
    if "cType" in args:
        cType = args["cType"]
    result =  0
    if cType == "*":
        result = 1
    for v in values:
        if cType == "+":
            result += v
        elif cType == "-":
            result -= v
        elif cType == "*":    
            result *= v
    return result
calculate(1,2,3,4,cType="*")
```
