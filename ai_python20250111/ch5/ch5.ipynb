```python
def myJoin(sp,myList):
    result = ""
    listLen = len(myList)
    for i in range(listLen):
        result += myList[i]
        if i <  listLen-1:
            result += sp
    return result         
sp = "-"
myList=["A","B","C","D"]            
s1 = myJoin(sp,myList)            
print(s1)
sp = "*"
s2 = myJoin(sp,myList)  
print(s2)
```

## 例外語法
try:
    保護的程式區段
    
except 例外1 :
    例外1發生時執行的程式
    
except :
    其他例外發生時執行的程式
    #未指定例外型態，捕捉所有例外物件
    
else:
    若try程式區段沒產生例外執行的程式
    
finally:
    不管有沒有發生例外都會執行的程式

```python
num1 = 10
num2 = 0
nums = [1,2,3,4]
print("Start")
try:
    print("X1")
    #print(num1 / num2)    
    print(nums[99])
    print("X2")
except:
    print("錯誤發生")
print("End")
```

```python
num1 = 10
num2 = 0
nums = [1,2,3,4]
print("Start")
try:
    print(nums[99])
    #print(num1/num2)
    #int("A")
except ZeroDivisionError:    
    print("ZeroDivisionError")
except IndexError:
    print("IndexError")
except:
    print("Error")
print("End")
```

```python
def testScore(score):
    if score > 100 or score < 0:
        print("不正確:Error")
        return
    print(f"成績:{score}")    
#    score必須0~100 正確顯示成績的數值  不正確 顯示Error
testScore(82)
testScore(250)
testScore(-15)
```

## 使用raise拋例外

```python
def testScore(score):
    if score > 100 or score < 0:
        raise OverflowError
    print(f"成績:{score}")    
#    score必須0~100 正確顯示成績的數值  不正確 顯示Error
testScore(82)
testScore(250)
testScore(-15)
```

## split 方法
+ split 方法能以比對規則針對字串進行切割後產生 list。
+ split 它的使用形式如下：
+ split(字串,最大分割次數)
   + 若不指定最大分割次數則全部分割

```python
str1 = "Ken,Vivin,Lucy"
myList = str1.split(",")
print(myList)
```

```python
import re
text = "one, two...ten"
re1 = re.split("[,. ]+",text)
print(re1)
```

## sub 方法
+ sub 方法用於替換，它的使用形式如下
  + sub(取代的字串或函數,來源字串,可設定被取代的數量)
+ 可設定被取代的數量若不指定時全部替換。

```python
import re
inputStr = "hello 123 world 456"
rStr = re.sub(r"\d+","222",inputStr)
print(rStr)
rStr = re.sub(r"\d+","222",inputStr,count=1)
print(rStr)
```
